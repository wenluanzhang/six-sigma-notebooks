# Run chart

The Run Chart is another useful tool, but instead of ascribing control limits, it identi
fies statistically significant patterns.

```python
import pandas as pd
import numpy as np
from scipy.stats import linregress
from stats_plots import imr_chart, run_chart
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
# Replace with your actual Excel path/sheet
xls = pd.ExcelFile("Six Sigma_Case Study.xlsx")
df = (
    pd.read_excel(xls, 
                  sheet_name="Control Charts")
        .dropna(how="all")
)
df.columns = df.columns.str.strip()
# Quick inspection of the data
df.head()
```

```python
# 3. Pass the Series directly to the function
imr_chart(df["Ex Caliper Mold Diameter IMR"], title="Ex Caliper Mold Diameter")
```

## Run chart

```python
run_chart(df, value_col="Ex Caliper Mold Diameter RunCh", chart_title="Run Chart for Ex Caliper Mold Diameter", ylabel="Ex Caliper Mold Diameter (mm)")
```

- The defects started to occur after the mold dimention increased to $(1/4)^"$ diameter.
- The root cause was determined to be faulty equipment used to de-slag of the mold.
- The mold diameter increased over time, which had an inverse effect of the Caliper Torsion Diameter.
- Here the trend of increasing mold size was confirmed.
