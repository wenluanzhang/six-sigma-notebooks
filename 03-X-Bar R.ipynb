# X-Bar R Chart

- A control chart that is recommended in lieu of the I-MR chart when the data is gathered in subgroups is the X-Bar R chart, which is for data that is in subgroups. This will allow us to also understand within-subgroup variation, which the I-MR would not do if we just plotted the average subgroup values.
- Use the X-Bar R chart when data is sampled in subgroups between 2 and 8 
samples. When subgroup sizes are 9 or greater, use the X-Bar S chart.

```python
import pandas as pd
import numpy as np
from stats_plots import xbarr_chart
import matplotlib.pyplot as plt
plt.style.use('classic')
```

## Load data

```python
# Replace with your actual Excel path/sheet
xls = pd.ExcelFile("Six Sigma_Case Study.xlsx")
df = pd.read_excel(xls, sheet_name="Control Charts", 
                   usecols=["Subgroup", "XBarR Caliper Torsion"]
                  ).copy().dropna(how='all')
df.columns = df.columns.str.strip()
# Quick inspection of the data
df.head()
```

```python
xbarr_chart(df, "Subgroup", "XBarR Caliper Torsion")
plt.show()
```
