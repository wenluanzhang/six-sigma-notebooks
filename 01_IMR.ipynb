# I-MR Control Chart: Brake Caliper Torsion
This notebook demonstrates how to create an Individuals & Moving Range (I-MR) control chart
for the "Brake Caliper Torsion IMR" dataset from the Six Sigma case study Excel file.
- The individual moving range control chart (I-MR) is the most versatile of control 
charts and can be used with most data types.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from stats_plots import imr_chart
plt.style.use('classic')
```

## Load data

```python
xls = pd.ExcelFile("Six Sigma_Case Study.xlsx")
df = (
    pd.read_excel(
        xls,
        sheet_name="Control Charts",
    )
    .dropna(how="all")  # drop rows where all columns are NaN
)
df.columns = df.columns.str.strip()
```

```python
# 2. Extract the column (as a pandas Series)
imr_data = df["Brake Caliper Torsion IMR"]

# 3. Pass the Series directly to the function
imr_chart(imr_data, title="Brake Caliper Torsion")

# 4. Show the plot
#plt.show()
```
