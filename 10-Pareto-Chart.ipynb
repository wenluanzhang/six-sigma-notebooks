# Pareto Chart
 A Pareto chart is a type of bar chart, where the X axis represents categories and the 
bars represent counts. The bars are arranged from highest to smallest counts. This 
allows us to look for situations where ~20% of the categories represent ~80% of the 
problem, and focus on what is important, or isolate areas where root causes 
might exist.

```python
import pandas as pd
import numpy as np
from stats_plots import pareto
import matplotlib.pyplot as plt
plt.style.use('classic')
```

```python
# Replace with your actual Excel path/sheet
xls = pd.ExcelFile("Six Sigma_Case Study.xlsx")
df = (
    pd.read_excel(xls, 
                   sheet_name="Pareto",)
        .dropna(how="all")
)
df.columns = df.columns.str.strip()
df.head()
```

```python
pareto(df, "Defect", title="Pareto Plot of Defects")
```

- Brakes and airbags account for 79.1% of all defects.
- This will carry forward to Analyze to get to the root cause, and will be the primary focus areas.
